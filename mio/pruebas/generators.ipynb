```python
def getfibo(n):
    i1 = 1
    i2 = 1
    for i in range(n):
        yield i2
        temp = i2
        i2 = i1
        i1 = i1 + temp

with open("fibo.txt", "w") as f:
    for a in getfibo(10000):
        f.write(str(a) + "\n")
```

vamos a definir una clase persona con el metodo call


```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    def saludar(self):
        print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} a침os.")
    def __str__(self):
        return f"Persona(nombre={self.nombre}, edad={self.edad})"
    def __call__(self, numero1, numero2, funcion):
        return f"Soy una persona que realiza una operaci칩n: {funcion(numero1, numero2)}"

p = Persona("Ana", 30)
p.saludar()
print(p)
resultado = p(5, 3, lambda x, y: x + y)
print(resultado)
```
