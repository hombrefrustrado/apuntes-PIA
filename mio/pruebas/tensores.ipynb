```python
import torch
import numpy as np
import time
```

```python
usar_gpu = True
device = "cuda" if usar_gpu and torch.cuda.is_available() else "cpu"
print(f"device asignado {device}")

time1 = time.time()
size=1000
tensor = torch.rand(size,size,dtype=torch.float16)

tensor_gpu = tensor.to(device)
len(tensor_gpu)

time2 = time.time()
print(time2-time1)
```

```python

class dataset():
    tamanyo_mesa = 100
    tamanyo_silla = 10
    def __init__(self,mesas:torch.tensor,sillas:torch.tensor):
        self.mesas = mesas
        self.sillas = sillas
    def __len__(self):
        return len(self.mesas) * dataset.tamanyo_mesa**2 + len(self.sillas) * dataset.tamanyo_silla**2
    
restaurante = dataset(torch.rand(10),torch.rand(32))
len(restaurante)
```
